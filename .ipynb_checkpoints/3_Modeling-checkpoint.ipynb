<a id="1.1"></a>
### 1.1 ARIMA Grid search:

```python
# Starting AIC, p, and q.
best_aic = 99 * (10 ** 16)
best_p = 0
best_q = 0

# Use nested for loop to iterate over values of p and q.
for p in range(10):
    for q in range(10):
        
        # Insert try and except statements.
        try:
            
            # Fitting an ARIMA(p, 0, q) model.
            print(f'Attempting to fit ARIMA({p},0,{q})')
            
            # Instantiate ARIMA model.
            arima = ARIMA(endog = y_train.astype(float).dropna(), # endog = Y variable
                          order = (p,0,q)) # values of p, d, q
            
            # Fit ARIMA model.
            model = arima.fit()

            # Print out AIC for ARIMA(p, 0, q) model.
            print(f'The AIC for ARIMA({p},0,{q}) is: {model.aic}')

            # Is my current model's AIC better than our best_aic?
            if model.aic < best_aic:
                
                # If so, let's overwrite best_aic, best_p, and best_q.
                best_aic = model.aic
                best_p = p
                best_q = q

        except:
            pass
print()
print()
print('MODEL FINISHED!')
print(f'Our model that minimizes AIC on the training data is the ARIMA({best_p},0,{best_q}).')
print(f'This model has an AIC of {best_aic}.')
```

MODEL FINISHED! Our model that minimizes AIC on the training data is the ARIMA(8,0,6). This model has an AIC of 4888.262727928628. Let's run an ARIMA model with best parameters and explore RMSE.

```python
arima = ARIMA(endog = y_train,
             order = (8,0,6))

model = arima.fit()
```

```python
model.aic
```

```python
pred = model.predict(start=y_test.index[0], end = y_test.index[-1])
```

```python
metrics.mean_squared_error(y_test, pred, squared=False)
```

```python
# plot all the data
ax = df['no_definite_rooms'].plot(label='Observed', alpha=0.75, figsize=(16,9))

# plot predicted values
pred.to_frame(name='Predicted').plot(ax=ax, label='Predicted', color='orange')

# title, x/y labels
plt.title('Rooms Sold with Prediceted Values ARIMA Model')
ax.set_xlabel('Year')
ax.set_ylabel('Rooms Sold')
plt.legend()
plt.show();
```

Grid search helped improve the RMSE by 3% from 27.83 to 27.11, which beats the hotel's 27.4 error.

### 3. Facebook's Prophet model - Rooms Sold:

Prophet is a procedure for forecasting time series data based on an additive model where non-linear trends are fit with yearly, weekly, and daily seasonality, plus holiday effects. It works best with time series that have strong seasonal effects and several seasons of historical data. Prophet is robust to missing data and shifts in the trend, and typically handles outliers well.
Prophet is open source software released by Facebook’s Core Data Science team.

Creating a Prophet-style dataframe as prophet needs two columns: "ds" for datetime and "y" for the predicted variable.

```python
df_o = pd.read_csv("./data/cleaned_data.csv").drop('Unnamed: 0', axis=1)
```

```python
df_o['business_date'] = pd.to_datetime(df_o['business_date'])
```

```python
df_o = df_o.groupby('business_date').agg('sum')
```

```python
df_o.reset_index()
```

```python
df = df.drop(columns=[col for col in df.columns if col not in ['no_definite_rooms']]).reset_index()
```

Setting up the "cap" column to limit forecast to the capacity of the hotel (120).

```python
df.columns= ['ds','y']
df['cap'] = 120 # maximum rooms sold, physical capacity
df.head()
```

Identifying the end index of the train data: 

```python
df.index[df['ds'] == '2018-12-31'].tolist()
```

```python
df = df[:549] #setting training data to 2017-2018
```

National holidays that effect hotel demand are added to the model to further affect deamnd forecasting over those days. 

```python
m = Prophet(growth='logistic')
m.add_country_holidays(country_name='US')
m.fit(df)
```

```python
print(m.train_holiday_names)
```

```python
ax = df_o['no_definite_rooms'].plot(label='Observed', alpha=0.75, figsize=(16,9), color='green')
future = m.make_future_dataframe(periods=365)
future['cap'] = 120
fcst = m.predict(future)
fig = m.plot(fcst, ax=ax, figsize=(16,9),xlabel='Date', ylabel='Rooms Sold')
```

```python
fig2 = m.plot_components(fcst)
```

```python
df_cv = cross_validation(m, initial='180 days', period='90 days', horizon = '365 days')
```

```python
df_cv.head()
```

```python
df_p = performance_metrics(df_cv)
df_p.head()
```

```python
fig = plot_cross_validation_metric(df_cv, metric='rmse')
```

```python
np.mean(df_p['rmse'])
```

FB Prophets showed a 9% improvement in RMSE over ARIMA model. Since it is a new package, some time was spent researching its capabilities and implementation. The actual training took seconds. There is a grid search component to this method which is explored below.

<a id="3.1"></a>
### 3.1 Prophet Grid Search:

```python
param_grid = {  
    'changepoint_prior_scale': [0.0001, 0.001, 0.01, 0.1, 0.5],
    'seasonality_prior_scale': [0.001, 0.01, 0.1, 1.0, 10.0],
}

# Generate all combinations of parameters
all_params = [dict(zip(param_grid.keys(), v)) for v in itertools.product(*param_grid.values())]
rmses = []  # Store the RMSEs for each params here

# Use cross validation to evaluate all parameters
for params in all_params:
    m = Prophet(**params).fit(df)  # Fit model with given params
    df_cv = cross_validation(m, initial='180 days', period='90 days', horizon = '365 days')
    df_p = performance_metrics(df_cv, rolling_window=1)
    rmses.append(df_p['rmse'].values[0])

# Find the best parameters
tuning_results = pd.DataFrame(all_params)
tuning_results['rmse'] = rmses
print(tuning_results)
```

Grid search through additional parameters did not improve the model, so original parameters will be used.

### 4. LSTM - RNN for Rooms Sold

LSTMs are sequential neural networks that assume dependence between the observations in a particular series.

```python
df = pd.read_csv("./data/cleaned_data.csv",  index_col='business_date',  ).drop('Unnamed: 0', axis=1)
```

```python
df.index = pd.to_datetime(df.index)
```

```python
# resampling on a daily level
y = df.resample('D')['no_definite_rooms'].sum()
y
```

```python
# normalize dataset with MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
df = scaler.fit_transform(y.values.reshape(-1, 1))
df
```

```python
# splitting train and test by index 
y_train, y_test = df[:549], df[549:914]
```

```python
# generating 7-day windows to be used for forecasting
train_seq = TimeseriesGenerator(y_train, y_train,
                                length=7, stride=1,
                                batch_size=3)
test_seq = TimeseriesGenerator(y_test, y_test,
                               length=7, stride=1,
                               batch_size=3)
```

```python
batch_x, batch_y = train_seq[0]
```

```python
batch_x.shape
```

```python
batch_y.shape
```

```python
input_shape = batch_x.shape[1:]
```

```python
model = keras.Sequential()
model.add(LSTM(32, input_shape=input_shape, return_sequences=False)) # started with 4 neurons 
model.add(Dense(1))
```

```python
model.summary()
```

```python
model.compile(loss='mean_squared_error',
              optimizer='adam',
              metrics=[tf.keras.metrics.RootMeanSquaredError()])
```

```python
history = model.fit(train_seq, 
                    validation_data=test_seq,
                    epochs=300,
                    verbose=0
                   )
```

```python
history = pd.DataFrame(history.history)
history.head()
```

```python
plt.figure(figsize=(16,9))
plt.plot(history['root_mean_squared_error'], label='train')
plt.plot(history['val_root_mean_squared_error'], label='test')
plt.legend();
```

This graph was heavily used to determine best direction to tweak the layout of RNN by looking at the trend for testing score.

```python
preds = scaler.inverse_transform(model.predict(test_seq))
```

```python
preds_train = scaler.inverse_transform(model.predict(train_seq))
```

```python
# calculate train RMSE
trainScore = math.sqrt(mean_squared_error(y[:542], preds_train[:,0]))
print('Train Score: %.2f RMSE' % (trainScore))
```

```python
# calculate test RMSE
testScore = math.sqrt(mean_squared_error(y[542:900], preds[:,0]))
print('Test Score: %.2f RMSE' % (testScore))
```

Baseline LSTM RNN did not permorm as well as other model with RMSE of 29.98 (worst out of all models so far) for the RNN with 32 neurons in the LSTM hidden layer and trained for 100 epochs. This best score was achieved via manual grid search through layer complexity and number of epochs. Automated grid search is recommended to explore tuning the RNN further if/when there are additional resources available.

```python
preds_df = pd.DataFrame(preds)
preds_df['Date'] = pd.date_range("2019-01-01", periods=358, freq="D")
preds_df.index = pd.to_datetime(preds_df['Date'])
```

```python
preds_df = preds_df.drop(['Date'], axis=1)
```

```python
ax = y.to_frame().plot(title='Observed', alpha=0.75, figsize=(16,9))
preds_df.plot(ax=ax, title='Dynamic Forecast', figsize=(16, 9), color='orange')
ax.set_xlabel('Date')
ax.set_ylabel('Rooms Sold')
plt.title('LSTM RNN for Rooms Sold')
plt.legend(labels=['Observed','Predicted'])
plt.show()
```

## 4.1 GRU - RNN Model Rooms

The GRU is the newer generation of Recurrent Neural Networks and is pretty similar to an LSTM. GRU is computationally more efficient due to less complex structure - GRU has just two gates: reset and update vs. LSTM having three gates: input, output and forget. This makes it a worthwhile candidate for this project's data.

```python
model = keras.Sequential()
model.add(GRU(64, input_shape=input_shape, return_sequences=False)) # started with 4 neurons 
model.add(Dense(1))
```

```python
model.summary()
```

```python
model.compile(loss='mean_squared_error',
              optimizer='adam',
              metrics=[tf.keras.metrics.RootMeanSquaredError()])
```

```python
history = model.fit(train_seq, 
                    validation_data=test_seq,
                    epochs=50,
                    verbose=0
                   )
```

```python
history = pd.DataFrame(history.history)
history.head()
```

```python
plt.figure(figsize=(16,9))
plt.plot(history['root_mean_squared_error'], label='train')
plt.plot(history['val_root_mean_squared_error'], label='test')
plt.legend();
```

The train/test score graph was used to manually grid search for best parameters (number of epochs and number of neurons in the GRU layer).

```python
preds = scaler.inverse_transform(model.predict(test_seq))
```

```python
preds.shape
```

```python
preds_train = scaler.inverse_transform(model.predict(train_seq))
```

```python
# calculate train RMSE
trainScore = math.sqrt(mean_squared_error(y[:542], preds_train[:,0]))
print('Train Score: %.2f RMSE' % (trainScore))
```

```python
# calculate test RMSE
testScore = math.sqrt(mean_squared_error(y[542:900], preds[:,0]))
print('Test Score: %.2f RMSE' % (testScore))
```

```python
preds_df = pd.DataFrame(preds)
preds_df['Date'] = pd.date_range("2019-01-01", periods=358, freq="D")
preds_df.index = pd.to_datetime(preds_df['Date'])
preds_df = preds_df.drop(['Date'], axis=1)
```

```python
ax = y.to_frame().plot(title='Observed', alpha=0.75, figsize=(16,9))
preds_df.plot(ax=ax, title='Dynamic Forecast', figsize=(16, 9), color='orange')
ax.set_xlabel('Date')
ax.set_ylabel('Rooms Sold')
plt.title('LSTM RNN for Rooms Sold')
plt.legend(labels=['Observed','Predicted'])
plt.show()
```

The best RMSE score of 29.87 was achieved with 8-neuron GRU hidden layer and training for 100 epochs. This score is lower than the baseline hotel forecast error and this model did not perform well with our data.

### 8.1 LSTM RNN for Revenue

```python
df = pd.read_csv("./data/cleaned_data.csv",  index_col='business_date',  ).drop('Unnamed: 0', axis=1)
```

```python
df.index = pd.to_datetime(df.index)
```

```python
y = df.resample('D')['revenue'].sum()
y
```

```python
# normalize dataset with MinMax Scaler since we don't want to forecast neative values 
scaler = MinMaxScaler()
df = scaler.fit_transform(y.values.reshape(-1, 1))
df
```

```python
y_train, y_test = df[:549], df[549:914] # splitting by index correspoding to 2019-12-31
```

```python
train_seq = TimeseriesGenerator(y_train, y_train,
                                length=7, stride=1, # length of 7 to capture full week of observations
                                batch_size=3)
test_seq = TimeseriesGenerator(y_test, y_test,
                               length=7, stride=1,
                               batch_size=3)
```

```python
batch_x, batch_y = train_seq[0]
```

```python
batch_x.shape
```

```python
input_shape = batch_x.shape[1:]
```

```python
model = keras.Sequential()
model.add(LSTM(8, input_shape=input_shape))
model.add(Dense(1))
```
